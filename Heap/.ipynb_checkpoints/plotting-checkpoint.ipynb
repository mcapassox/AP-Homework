```python
x = [10,100,1000,10000, 25000,50000]
recursive = [1E-7,1E-06,1.4E-05,0.000116,0.000257,0.000526]
iterative = [1E-7,1E-06,9E-06,0.000104,0.000264,0.000516]
```
