```python
x = [1000,10000,25000,50000]
insertion_worst = [0.000512,0.040138,0.248151,0.990706]
insertion_best = [3E-06,1.4E-05,3.6E-05,5.9E-05]
insertion_best_fit = zeros(4)
insertion_worst_fit = zeros(4)

z = np.polyfit(x, insertion_worst, 2)

counter = 0

for i in x:
    insertion_worst_fit[counter] = z[2]+z[1]*i+z[0]*i*i
    counter += 1
    
z = np.polyfit(x, insertion_best, 1)
print(z)

counter = 0
for i in x:
    insertion_best_fit[counter] = z[1]+z[0]*i
    counter += 1
```

```python
semilogy(x, insertion_worst, 'o')
plot(x, insertion_worst_fit)

plot(x, insertion_best, '.')
plot(x, insertion_best_fit)



#loglog(x, quick_best)
ylabel('Running time (s) ')
xlabel('Size of Array')
title('Running time')
legend(['insertion (WC)', 'insertion_quad_fit', 'insertion (BC)', 'insertion_lin_fit'])
savefig('running_time_bwinsertion.pdf')
```

```python
x = [1000,10000,25000,50000]
quick_worst = [0.000525,0.036685,0.262399,0.990039]
quick_best = [2.4E-05,0.00018,0.000598,0.001387]
quick_best_fit = zeros(4)
quick_worst_fit = zeros(4)

z = np.polyfit(x, quick_worst, 2)

counter = 0

for i in x:
    quick_worst_fit[counter] = z[1]*i+z[0]*i*i
    counter += 1

z = np.polyfit(numpy.log(x)*(x), quick_best, 1)

counter = 0
for i in x:
    quick_best_fit[counter] = z[0]*log(i)*(i)
    counter += 1

semilogy(x, quick_worst, 'o')
plot(x, quick_worst_fit)

plot(x, quick_best, '.')
plot(x, quick_best_fit)

print()

#loglog(x, quick_best)
ylabel('Running time (s) ')
xlabel('Size of Array')
title('Running time')
legend(['quick sort (WC)', 'quick_quad_fit', 'quick sort (BC)', 'quick_log_fit'])
savefig('running_time_bwquick.pdf')
```
