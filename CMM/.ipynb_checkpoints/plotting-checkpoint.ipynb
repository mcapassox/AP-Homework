```python
%pylab inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
x = [10,50,100,200,500,700,1000]
optimal_constn = [1.36E-04, 8.71E-04, 5.52E-03, 5.97E-02, 5.98E-01, 1.77E+00, 4.62E+00]
naive_constn = [2.00E-05, 1.35E-03, 1.21E-02, 1.34E-01, 1.73E+00, 1.83E+00, 6.40E+00]
```

```python
plot(x,optimal_constn)
plot(x,naive_constn)


ylabel('Time (s) ')
xlabel('Maximum matrix size')
title('Running time (n = 50)')

legend(['Optimal', 'Naive'])
savefig('const_n.pdf')
```
